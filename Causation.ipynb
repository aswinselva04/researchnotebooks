## Granger Causation

```python
import pandas as pd
```

```python
overall_data = pd.read_excel('tfcu_member.xlsx', index_col=0, sheet_name='KPI Data')
df_ = overall_data.drop(overall_data.index[0:3])
df = df_.drop(columns=['Unnamed: 10'])
df_time = df[['Week','Member Retention',
       'Employee Retention', 'Member Satisfaction NPS', 'Fraud rate',
       'Employee satisfaction', 'Number of Employee Relations Issues',
       '# of Member complaints', 'Member acquisition cost (CAC)',
       'Number of Compliance Issues', 'Compliance Investigation Cycle Time',
       'Decline rate', 'Financial impact due to attacks',
       'Attempted vs Fraudulent Transactions', 'Final Approval rate',
       'Fraudulent Vs Declined Transactions', 'System Uptime',
       'Severity of attacks', 'Mean Time Between Failures (MTBF)',
       'Cost per incident', 'Number of cybersecurity incidents reported'
             ]].groupby(['Week']).mean()
```

```python
df_time.head()
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
grangercausalitytests(df_time[['Employee Retention', 'Compliance Investigation Cycle Time']].diff().dropna(), maxlag=10)
```

```python
grangercausalitytests(df_time[[ 'Compliance Investigation Cycle Time', 'Employee Retention']], maxlag=4)
```

```python
from darts.timeseries import TimeSeries
```

```python
df_correlated = df_time['Employee Retention'].reset_index().drop('Week', axis=1) * 2
```

```python
from darts.utils.statistics import granger_causality_tests

ts_cause = TimeSeries.from_series(df_time['Employee Retention'].reset_index().drop('Week', axis=1))
# ts_effect = TimeSeries.from_series(df_time['Compliance Investigation Cycle Time'].reset_index().drop('Week', axis=1))
ts_effect = TimeSeries.from_series(df_correlated)
maxlag = 4
granger_causality_tests(ts_cause, ts_effect, maxlag, addconst=True)
```

## Google Causal Impact

```python
from causalimpact import CausalImpact
import numpy as np
from statsmodels.tsa.arima_process import arma_generate_sample
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x1 = arma_generate_sample(ar=[0.999], ma=[0.9], nsample=100) + 100
y = 1.2 *x1 + np.random.randn(100)
```

```python
y[71:100] = y[71:100] + 10
```

```python
data = pd.DataFrame(np.array([y, x1]).T, columns=["y","x1"])
pre_period = [0,69]
post_period = [71,99]
```

```python

pre_period = [1, 45]
post_period = [46, 52]
```

```python
# impact = CausalImpact(df_time[['Employee Retention', 'Compliance Investigation Cycle Time']], pre_period, post_period)
impact =  CausalImpact(data, pre_period, post_period)
```

```python
impact.run()
```
